```python
import pandas as pd
import numpy as np
```

```python
cleanedmath_df = pd.read_csv('data/cleanedStudentMath.csv')
cleanedpor_df = pd.read_csv('data/cleanedStudentPor.csv')
```

```python
math_df = pd.read_csv('data/student-mat.csv')
por_df = pd.read_csv('data/student-por.csv')
```

```python
math_df = math_df[math_df.columns[0]].str.split(';', expand=True)
por_df = por_df[por_df.columns[0]].str.split(';', expand=True)
```

```python
columnLabels = ['School', 'Sex', 'Age', 'Address', 'Family Size', 'Parental Status', 'Mother Education'
             , 'Father Education', 'Mother Job', 'Father Job', 'Reason for School', 'Guardian', 'Travel Time'
             , 'Study Time', '# of Failures', 'Extra Educational Support', 'Extra Family Support', 'Extra Paid Classes'
             , 'Extracurricular', 'Nursery', 'Desire Higher Education', 'Internet Access', 'Romantic Relationship'
             , 'Family Relationship Quality', 'Amount of Free Time', 'Frequency of Going Out', 'Weekday Alcohol Consumption'
             , 'Weekend Alcohol Consumption', 'Health Status', 'Abscences', '1st Period Grade', '2nd Period Grade'
             , 'Final Grade']
math_df.columns = columnLabels
por_df.columns = columnLabels
```

```python
for x in math_df.columns:
    math_df[x] = math_df[x].str.strip("\"")
```
