```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
math2_df = pd.read_csv("data/math_data.csv")
```

```python
math2_df
```

```python
math2_data_ex = math2_df.iloc[:, :-3] # Only taking features that isn't the final grade or semester grades
math2_data_inc = math2_df.iloc[:, :-1] # Only taking features that isn't the final grade
math2_outcome = math2_df.iloc[:, -1:] # Only want the outcome of the final grade.
math2_protected = math2_df.drop(columns = ['Sex', 'Age', 'Mother Education', 'Father Education', 'Mother Job', 'Father Job', 'Guardian', 'Parental Status'])
```

```python
# Preprocessing of math2 data df to turn binary string values and categorial variables into integers
# Copying excluded and included datasets to keep original intact
no_yes_answer = {'no': 0,'yes': 1}
school_identity = {'GP': 0,'MS': 1} 
urban_rural = {'U': 0,'R': 1} 
family_size = {'LE3': 0,'GT3': 1}  
binary_categories = {
                     "Extra Educational Support": no_yes_answer,
                     "Extra Family Support": no_yes_answer,
                     "Extra Paid Classes": no_yes_answer,
                     "Extracurricular": no_yes_answer,
                     "Nursery": no_yes_answer,
                     "Desire Higher Education": no_yes_answer,
                     "Internet Access": no_yes_answer,
                     "Romantic Relationship": no_yes_answer,
                     "School": school_identity,
                     "Address": urban_rural,
                     "Family Size": family_size,
                    }

encode_math2_data_ex = math2_data_ex.copy()
encode_math2_data_inc = math2_data_inc.copy()

encode_math2_data_ex = encode_math2_data_ex.replace(binary_categories)
encode_math2_data_inc = encode_math2_data_inc.replace(binary_categories)

categorial_features = encode_math2_data_ex.select_dtypes(include=object).columns.tolist()
encode_math2_data_ex = pd.get_dummies(encode_math2_data_ex, columns=categorial_features)
encode_math2_data_inc = pd.get_dummies(encode_math2_data_inc, columns=categorial_features)
```

```python
# Preprocessing of math2 outcome df to turn final grade values into a binary pass/no pass
# Making a copy of the df since we want to keep the original dataframe
def bin_conversion(outcomes):
    for i, score in enumerate(outcomes["Final Grade"]):
        if score > 9:
            outcomes.at[i] = 1
        else:
            outcomes.at[i] = 0
    
    return outcomes

bin_math2_outcome = math2_outcome.copy()
bin_math2_outcome = bin_conversion(bin_math2_outcome)
```

```python
# Splitting one hot encoded data and binary math2 outcomes into train and test sets
# This cell is creation of train and test datasets that excludes semester grades
from sklearn.model_selection import train_test_split
math2_data_train_ex, math2_data_test_ex, math2_outcome_train_ex, math2_outcome_test_ex = train_test_split(encode_math2_data_ex, bin_math2_outcome, test_size=0.30,shuffle=True)
math2_data_train_ex, math2_data_test_ex, math2_outcome_train_ex, math2_outcome_test_ex = math2_data_train_ex.values, math2_data_test_ex.values, math2_outcome_train_ex.values.ravel(), math2_outcome_test_ex.values.ravel()
```

```python
# This cell is creation of train and test datasets that includes semester grades
math2_data_train_inc, math2_data_test_inc, math2_outcome_train_inc, math2_outcome_test_inc = train_test_split(encode_math2_data_inc, bin_math2_outcome, test_size=0.30,shuffle=True)
math2_data_train_inc, math2_data_test_inc, math2_outcome_train_inc, math2_outcome_test_inc = math2_data_train_inc.values, math2_data_test_inc.values, math2_outcome_train_inc.values.ravel(), math2_outcome_test_inc.values.ravel()
```

```python
def mean_squared_error(y, y_hat):
    from sklearn.metrics import mean_squared_error
    return mean_squared_error(y, y_hat)
```

```python
def correct_percentage(y, y_hat):
    n = len(y)
    accuracyList = []
    for i in range(n):
        if y[i] == y_hat[i]:
            accuracyList.append(1)
        else:
            accuracyList.append(0)
            
    return (sum(accuracyList) / n) * 100
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}%'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
# Using four different machine learning models on data
# 1. Decision Tree
# 2. Random Forest
# 3. Ridge Regression
# 4. Neural Net
from sklearn import linear_model, tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
dt = tree.DecisionTreeClassifier(max_depth=5)
rf = RandomForestClassifier(max_depth=5)
ridgereg = linear_model.Ridge()
neurnet = MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=500)
scaler = StandardScaler()
```

```python
# Next few cells will be fitting and prediciting the models on the exlucded dataset
# Decision tree model
dt.fit(math2_data_train_ex, math2_outcome_train_ex)
dt_yhat = dt.predict(math2_data_test_ex)
dt_per_binex = correct_percentage(math2_outcome_test_ex, dt_yhat)
```

```python
# Random forest model
rf.fit(math2_data_train_ex, math2_outcome_train_ex)
rf_yhat = rf.predict(math2_data_test_ex)
rf_per_binex = correct_percentage(math2_outcome_test_ex, rf_yhat)
```

```python
# Ridge regression model
ridgereg.fit(math2_data_train_ex, math2_outcome_train_ex)
ridgereg_yhat = ridgereg.predict(math2_data_test_ex)
ridgereg_per_binex = correct_percentage(math2_outcome_test_ex, np.round(ridgereg_yhat))
```

```python
# Neural network model, input data needs to be scaled in order to fit data
scaler.fit(math2_data_train_ex)
math2_data_train_scaled = scaler.transform(math2_data_train_ex)
math2_data_test_scaled = scaler.transform(math2_data_test_ex)
neurnet.fit(math2_data_train_scaled, math2_outcome_train_ex)
neurnet_yhat = neurnet.predict(math2_data_test_scaled)
neurnet_per_binex = correct_percentage(math2_outcome_test_ex, neurnet_yhat)
```

```python
# Next few cells will be fitting and prediciting the models on the included dataset
# Decision tree model
dt.fit(math2_data_train_inc, math2_outcome_train_inc)
dt_yhat = dt.predict(math2_data_test_inc)
dt_per_bininc = correct_percentage(math2_outcome_test_inc, dt_yhat)
```

```python
# Random forest model
rf.fit(math2_data_train_inc, math2_outcome_train_inc)
rf_yhat = rf.predict(math2_data_test_inc)
rf_per_bininc = correct_percentage(math2_outcome_test_inc, rf_yhat)
```

```python
# Ridge regression model
ridgereg.fit(math2_data_train_inc, math2_outcome_train_inc)
ridgereg_yhat = ridgereg.predict(math2_data_test_inc)
ridgereg_per_bininc = correct_percentage(math2_outcome_test_inc, np.round(ridgereg_yhat))
```

```python
# Neural network model, input data needs to be scaled in order to fit data
scaler.fit(math2_data_train_inc)
math2_data_train_scaled = scaler.transform(math2_data_train_inc)
math2_data_test_scaled = scaler.transform(math2_data_test_inc)
neurnet.fit(math2_data_train_scaled, math2_outcome_train_inc)
neurnet_yhat = neurnet.predict(math2_data_test_scaled)
neurnet_per_bininc = correct_percentage(math2_outcome_test_inc, neurnet_yhat)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(11,6))
bin_ex_percentages = np.round([dt_per_binex, rf_per_binex, ridgereg_per_binex, neurnet_per_binex], 2)
bin_inc_percentages= np.round([dt_per_bininc, rf_per_bininc, ridgereg_per_bininc, neurnet_per_bininc], 2)
x_labels = ["Decision Tree", "Random Forest", "Ridge Regression", "Neural Net"]

barWidth = 0.25

x = np.arange(len(x_labels))
plt.xticks(x, x_labels)

r1 = plt.bar(x - barWidth/2, bin_ex_percentages, width=barWidth, color="red", label="Excluding Semester Grades")
r2 = plt.bar(x + barWidth/2, bin_inc_percentages, width=barWidth, color="blue", label="Including Semester Grades")

plt.ylabel("Accuracy Percentage")
plt.xlabel("Models", fontweight="bold")
plt.ylim(top=100, bottom=0)
plt.title("Accuracy percentage of models with binary outcomes (math_protected)")
plt.legend()
plt.savefig('math_protected.png')
plt.show()
```
